```python
#IMPORTANT FEATURES REQUIRED FOR MUSIC RECOMMENDATION ENGINE ARE COMBINED
feature=['Songname','Singer','Genre','Movie']
bollywood_movie[feature]
```

```python
#FUNCTION THAT COMBINE ALL THE IMPORTANT FEATURES INTO A SINGLE STRING 
def factor_determiner(data):
    feature=[]
    for i in range(0,data.shape[0]):
        feature.append(data['Songname'][i]+','+data['Singer'][i]+','+data['Genre'][i]+','+data['Movie'][i])
    return feature
```

```python
#ADD THE COMBINED STRING AS A NEW COLUMN TO THE DATASET
bollywood_movie['feature']=factor_determiner(bollywood_movie)
#PRINT FIRST 5 DETAILS
bollywood_movie.head(5)
```

```python
#Import libarries to convert the text(feature) into token counts
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
cosinematrix=CountVectorizer().fit_transform(bollywood_movie['feature'])
similaritymatrix=cosine_similarity(cosinematrix)
#Print the similarity matrix
similaritymatrix
```

# BASED ON THE USERS INPUT TRY TO RECOMMEND THE SIMILAR SONGS 


```python
#User searches song based on genre
genre="BollywoodSad"
#Find the songsid
songid=bollywood_movie[bollywood_movie.Genre == genre]['Songid'].values[0]
#Create a list of enumerations for similarity scores
matchingscores=list(enumerate(similaritymatrix[songid]))
#Print the matchingscores to given genre
#Values are in the range of 0-1 where values near 1 indicate the most matched songid and to 0 are least matched songid
matchingscores
```

```python
#Sort the matching similarity scores in the descending order
sorted_scores=sorted(matchingscores,key=lambda x:x[1],reverse=True)
#Dont wnat the same song itself
sorted_scores=sorted_scores[1:]
#print the sorted matching scores
sorted_scores
```

```python
#Create a function which generates top ten songnames matching the given attributes
j=0
print("Recommended songs as per your",genre,"are:\n")
for item in sorted_scores:
    song_name=bollywood_movie[bollywood_movie.Songid==item[0]]['Songname'].values[0]
    print(j+1, song_name)
    j=j+1
    if(j>9):
        break
```
